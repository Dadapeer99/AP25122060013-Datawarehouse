# 10.Binomial Distribution & PMF
### PMF → binom.pmf(k, n, p) → exact probability of k successes.
### CDF → binom.cdf(k, n, p) → probability of ≤ k successes.
### SF → binom.sf(k, n, p) → probability of > k successes.***


```python
import numpy as np
A=np.array([[1,2],
            [3,4]])
B=np.array([[5,6],
            [7,8]])
```

```python
# Probability (approx by many experiments)
import math
from scipy.stats import binom
p = binom.pmf(500, 1000, 0.5)
print("P(500 heads) =", p)

# It models the probability of getting exactly k successes in n independent trials,
# P(X=6) when n=10, p=0.5
prob = binom.pmf(6, n=10, p=0.5)
print(prob)  # 0.205078125
```
